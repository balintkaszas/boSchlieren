```python
import numpy as np
#import cv2
framerate = 29.18
from scipy import signal
from scipy.signal import argrelextrema
from scipy.optimize import curve_fit

import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_widths
#%cd "D:\Interfacial Waves resonance videos 0430"
framerate = 29.18
```

```python
def lin(x,a,b):
    return a*x+b
```

```python
driving = np.loadtxt('data0523/1_akad_ramp_fel_0_75_forcing.dat', skiprows=2)
fr = 1/(driving[1,0]-driving[0,0]) ##framerate
from scipy import signal
N  = int(20*fr)
f, t, Zxx = signal.stft(driving[:,1], fr, nperseg=4*N,noverlap=4*N-1, detrend='constant')
freqs = f[np.argmax(np.abs(Zxx), axis=0)]
popt, pcov = curve_fit(lin, t[N:-N], freqs[N:-N])
error = np.sqrt(np.diag(pcov))
plt.plot(t, freqs, 'o')
plt.plot(t, lin(t, *popt), '-', c='Crimson', linewidth=2.)
print("m = ", popt[0], " +/- ", error[0])
print("offset = ", popt[1], " +/- ", error[1])
print("Df/dT = ", (freqs[-1]-freqs[0])/driving[:,0][-1])
```
